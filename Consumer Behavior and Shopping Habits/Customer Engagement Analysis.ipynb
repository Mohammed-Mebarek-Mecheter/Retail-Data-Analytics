```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("data/shopping_behavior.csv")
```

```python
# Print the first 5 rows of the dataset
print(data.head())
```

### Review Ratings

```python
# Plot the distribution of review ratings provided by customers
plt.figure(figsize=(8, 6))
sns.countplot(x='Review Rating', data=data, order=data['Review Rating'].value_counts().index, hue='Review Rating', legend=False)
sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
plt.title('Distribution of Review Ratings')
plt.xlabel('Review Rating')
plt.ylabel('Frequency')
plt.show()
```

### Subscription Status

```python
# Calculate the proportion of customers with subscription status
subscription_proportion = data['Subscription Status'].value_counts(normalize=True) * 100

# Display the proportion of customers with subscription services
print("Proportion of customers with subscription services:")
print(subscription_proportion)
```

```python
# Plot the distribution of subscription status
plt.figure(figsize=(8, 6))
sns.countplot(x='Subscription Status', data=data, order=data['Subscription Status'].value_counts().index, hue='Subscription Status', legend=False)
sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
plt.title('Distribution of Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate average previous purchases and spending for customers with and without subscription status
average_previous_purchases = data.groupby('Subscription Status')['Previous Purchases'].mean()
average_purchase_amount = data.groupby('Subscription Status')['Purchase Amount (USD)'].mean()

# Display the average purchase frequency and spending for customers with and without subscription status
print("Average previous purchases for customers with and without subscription status:")
print(average_previous_purchases)

print("\nAverage purchase amount (USD) for customers with and without subscription status:")
print(average_purchase_amount)
```

```python
# plot the average previous purchases and spending for customers with and without subscription status
plt.figure(figsize=(8, 6))
sns.barplot(x=average_previous_purchases.index, y=average_previous_purchases.values, hue=average_previous_purchases.index, legend=False)
sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
plt.title('Average Previous Purchases')
plt.xlabel('Subscription Status')
plt.ylabel('Frequency')
plt.show()
```

### Frequency of Purchases

```python
# Calculate the average frequency of purchases
average_purchase_frequency = data['Frequency of Purchases'].value_counts()

# Display the average frequency of purchases
print("Average frequency of purchases:", average_purchase_frequency)
```

```python
# Plot the distribution of frequency of purchases
plt.figure(figsize=(8, 6))
sns.countplot(x='Frequency of Purchases', data=data, order=data['Frequency of Purchases'].value_counts().index, hue='Frequency of Purchases', legend=False)
sns.color_palette("mako", as_cmap=True)
plt.title('Distribution of Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate average frequency of purchases and spending for customers with and without subscription status
average_purchase_frequency = data.groupby('Subscription Status')['Frequency of Purchases'].value_counts().mean()
average_purchase_amount = data.groupby('Subscription Status')['Purchase Amount (USD)'].mean()

# Display the average purchase frequency and spending for customers with and without subscription status
print("Average previous purchases for customers with and without subscription status:")
print(average_purchase_frequency)

print("\nAverage purchase amount (USD) for customers with and without subscription status:")
print(average_purchase_amount)
```

### Promotional Offers

```python
# Calculate the frequency of engagement with promotional offers
promotional_offer_engagement = data['Discount Applied'].value_counts(normalize=True) * 100

# Display the frequency of engagement with promotional offers
print("Frequency of engagement with promotional offers:")
print(promotional_offer_engagement)
```

```python
# Plot the distribution of engagement with promotional offers
plt.figure(figsize=(8, 6))
sns.countplot(x='Discount Applied', data=data, order=data['Discount Applied'].value_counts().index, hue='Discount Applied', legend=False)
sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
plt.title('Distribution of Engagement with Promotional Offers')
plt.xlabel('Engagement with Promotional Offers')
plt.ylabel('Frequency')
plt.show()
```
